```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from sklearn.metrics import f1_score
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
```

```python
data = pd.read_csv('/content/drive/MyDrive/Generative_AI_Dataset/train.csv')
test= pd.read_csv('/content/drive/MyDrive/Generative_AI_Dataset/test.csv')
sample_sub= pd.read_csv('/content/drive/MyDrive/Generative_AI_Dataset/solution_format.csv')
```

```python
data
```

```python
# Split the data into features and labels
features = data.iloc[:, 1:]
labels = data.iloc[:, 0]
```

```python
features
```

```python
labels
```

```python
#normalizing
x = features.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
features_normalized = pd.DataFrame(x_scaled)
```

```python
features_normalized
```

```python
assert features_normalized.shape[1] == 20 * 20 * 3
```

```python
new_values = features_normalized.values
```

```python
features_reshaped = np.reshape(new_values, (5250, 20, 20, 3))
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_reshaped, labels, test_size=0.2, random_state=42)
```

```python
X_train_flattened = X_train.reshape(X_train.shape[0], -1)
X_test_flattened = X_test.reshape(X_test.shape[0], -1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA

# Apply PCA for dimensionality reduction
pca = PCA(n_components=185)
X_train_pca = pca.fit_transform(X_train_flattened)
X_test_pca = pca.transform(X_test_flattened)
# Create an AdaBoost classifier
model = AdaBoostClassifier()

# Train the model on the PCA-transformed data
model.fit(X_train_pca, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_pca)
y_pred = np.round(y_pred).astype(int)

# Calculate F1 score
f1 = f1_score(y_test, y_pred)
print('F1 score:', f1)
```

```python
# Preprocess the test data
test_features = test.iloc[:, 1:]
test_features_normalized = min_max_scaler.transform(test_features)
test_features_pca = pca.transform(test_features_normalized)

# Make predictions on the test data
test_predictions = model.predict(test_features_pca)
test_predictions = np.round(test_predictions).astype(int)
```

```python
# Create a submission DataFrame
submission = pd.DataFrame({'id': test['id'], 'labels': test_predictions})

# Save the submission to a CSV file
submission.to_csv('submission.csv', index=False)
```
